# Resumo
Este trabalho propõe uma metodologia para facilitar a compreensão dos demonstrativos financeiros de empresas por meio de um modelo de linguagem. Reconhece-se que a leitura desses relatórios é desafiadora para o público geral devido à complexidade, extensão e linguagem técnica. Para superar esses desafios, o método proposto utiliza um LLM (Large Language Model) personalizado, que processa os dados em etapas: carregamento, divisão e transformação do texto em representações numéricas, armazenando essas representações para consultas posteriores.
Ao acessar dados específicos, a metodologia permite uma abordagem eficaz, fornecendo respostas ágeis e simplificadas. No entanto, ressalta-se a necessidade de equilibrar a complexidade do modelo para evitar respostas imprecisas ao buscar detalhes mais elaborados.

# Descrição do problema

A leitura de demonstrativos financeiros de empresas pode ser um desafio para uma pessoa comum devido à complexidade dos documentos, a linguagem técnica utilizada, a natureza quantitativa dos dados, a falta de contexto, a extensão dos relatórios e a presença de riscos e incertezas. Superar esses obstáculos requer a busca de orientação profissional, o estudo de conceitos financeiros básicos e a utilização de fontes confiáveis de informação.

# Objetivo

Facilitar a compreensão e análise de demonstrativos financeiros de empresas por meio da utilização de um poderoso modelo de linguagem.

# Descrição da metodologia aplicada

Um demonstrativo financeiro de uma empresa é um documento em formato PDF que contém dezenas de páginas de informações detalhadas sobre a empresa. Geralmente, esse demonstrativo é elaborado anualmente. No entanto, criar um LLM (Large Language Model) personalizado a partir desses dados pode ser desafiador, uma vez que a maioria dos LLMs existentes não suporta o processamento de uma quantidade tão extensa de informações. Portanto, é necessário encontrar outra maneira de realizar a integração com os dados, inserindo apenas os dados essenciais para que o LLM possa fornecer uma resposta satisfatória ao prompt fornecido

## Inserção de dados

Para enviar apenas os dados necessários para uma solicitação específica, é necessário passar por um processo de preparação de dados dividido em quatro etapas:
1. **Carregamento de Fontes de Dados em Texto:** Nesta etapa, os dados são convertidos a partir de um formato que não é diretamente utilizável, como o formato PDF, para uma forma arbitrária de texto que pode ser posteriormente processada.
2. **Divisão do Texto em Segmentos:** O segundo passo envolve a divisão do texto carregado em segmentos menores. Isso é necessário devido às limitações de capacidade dos modelos de linguagem, que geralmente têm um limite de quantidade de texto que podem processar. Portanto, é essencial reduzir o texto em segmentos gerenciáveis.
3. **Transformação do Texto em Representações Numéricas (Embeddings):** Na terceira etapa, os segmentos de texto são transformados em valores numéricos, conhecidos como embeddings. Essas representações numéricas são essenciais para comparar a similaridade entre diferentes segmentos de texto. A finalidade é selecionar os trechos mais relevantes para uma pergunta específica, o que é alcançado encontrando os trechos mais semelhantes no espaço de embeddings.
4. **Armazenamento dos Embeddings no Vectorstore:** Na última etapa, os embeddings gerados anteriormente são salvos em um vectorstore. Os vectorstores são recursos que facilitam a busca eficiente de segmentos de texto mais similares no espaço de embeddings, tornando o processo rápido e eficaz.

## Consulta de dados

Após a divisão dos dados em segmentos mais gerenciáveis, e realizado o processo de consulta e geração de respota as pergunta feita pelo usuário. Esse processo pode ser exemplificado em 3 etapas:

1. **Unificação do Histórico de Conversa e da Nova Pergunta:** Nesta etapa, o histórico da conversa e a nova pergunta são combinados em uma única pergunta independente. Isso é fundamental para possibilitar a capacidade de fazer perguntas de acompanhamento, garantindo que o contexto da conversa seja mantido.

2. **Busca por Documentos Relevantes:** Utilizando os embeddings e o vectorstore criados durante a ingestão dos dados, realizamos uma pesquisa para identificar documentos relevantes que possam auxiliar na geração de respostas mais precisas e contextualmente sólidas. Isso envolve encontrar informações pertinentes que estão armazenadas nos documentos.

3. **Geração de Resposta:** Com a pergunta independente e os documentos relevantes à disposição, podemos empregar um modelo de linguagem para gerar a resposta à pergunta feita pelo usuário. Essa etapa permite que o sistema forneça uma resposta com base no contexto e nas informações disponíveis.

### Extraindo os dados do pdfs.

```python
path = "/content/llm-para-auxiliar-leitura-de-demonstrativos-financeiros/dados/demonstrativos-banco-do-nordeste"
list_file_names = os.listdir(path)


for file in list_file_names:
  doc = textract.process(f"{path}/{file}")
  file_name = file.replace('.pdf', '')
  with open(f'alldata.txt', 'a', encoding="utf-8") as f:
    f.write(doc.decode('utf-8'))
```

### Tokenizar e embeddings os dados

```python

with open('alldata.txt', 'r', encoding="utf-8") as f:
  text = f.read()

tokenizer = GPT2TokenizerFast.from_pretrained("gpt2")

def count_tokens(text: str) -> int:
  return len(tokenizer.encode(text))

text_splitter = RecursiveCharacterTextSplitter(
  chunk_size = 512,
  chunk_overlap  = 24,
  length_function = count_tokens,
  )

chunks = text_splitter.create_documents([text])

embeddings = OpenAIEmbeddings(openai_api_key=key_openai, model="text-embedding-ada-002")

db = FAISS.from_documents(chunks, embeddings)

vectorstoreData = db
```

```python
with open("vectorstore_demo-2010-2022.pkl", "wb") as f:
	pickle.dump(db, f)
```

### Parte 2
Carregando Vectorstore.

```python
with open("/content/vectorstore_demo-2010-2022.pkl", "rb") as f:
  vectorstoreData = pickle.load(f)

```

### Criando Modelo

```python
retriever = VectorStoreRetriever(vectorstore=vectorstoreData)

llm = OpenAI(openai_api_key=key_openai, temperature=0.1)
retriever = retriever
memory = ConversationBufferMemory(
    memory_key="chat_history", return_messages=True)
model = ConversationalRetrievalChain.from_llm(
    llm=llm,
    retriever=retriever,
    memory=memory)
```

### Realizar a pergunta

```python
question = "Como está os resultados financeiros do banco em 2022?"
```

```python
result = model({"question": question})
result['answer']
```
